```python
def hex_bin(string):
    return " ".join(
        reversed([i + j for i, j in zip(*[["{0:04b}".format(int(c, 16)) for c in reversed("0" + string)][n::2] for n
                                          in [1, 0]])]))

def toStr(s):
    return s and chr(int(s[:2], base=16)) + toStr(s[2:]) or ''


def bin_hex(string):
    return hex(int(string, 2))


def xor(a, b, c, d):
    e = hex(int(a,16) ^ int(b,16) ^ (int(c,16) if c != "" else 0) ^ (int(d,16) if d != "" else 0))
    return "0"+e[2:] if len(e[2:])==1 else e[2:]


def multi_xor(col1, col2, col3):
    new_col = []
    if isinstance(col3, list):
        for x,y,z in zip(col1, col2, col3):
            new_col.append(xor(x,y,z,""))
        return new_col
    else:
        for x,y in zip(col1, col2):
            new_col.append(xor(x,y,"",""))
        return new_col


def subdivide(matrix):
    dico = {}
    id_k = 0
    n_key = [[],[],[],[]]
    counter = 0
    for i in range(len(matrix[0])):
        counter += 1
        n_key[0].append(matrix[0][i])
        n_key[1].append(matrix[1][i])
        n_key[2].append(matrix[2][i])
        n_key[3].append(matrix[3][i])
        if counter == 4 and id_k < 15:
            dico["k"+str(id_k)] = n_key
            id_k += 1
            n_key = [[],[],[],[]]
            counter = 0
    return dico


def split_in(block, char):
    return [block[i:i + char] for i in range(0, len(block), char)]
```

```python
def galois_multiplication(a, b):
    """Galois multiplication of 8 bit characters a and b."""
    a = int(a,16)
    b = int(b,16)
    p = 0
    for counter in range(8):
        if b & 1: p ^= a
        hi_bit_set = a & 0x80
        a <<= 1
        # keep a 8 bit
        a &= 0xFF
        if hi_bit_set:
            a ^= 0x1b
        b >>= 1
    return str(hex(p))[2:]
```

```python
def subBytes(mat,):
    matsub = [[],[],[],[]]
    for i in range(len(mat)):
        line = mat[i]
        for j in line:
            x = j[:1]
            y = j[1:]
            matsub[i].append(SBOX[int(x,16)][int(y,16)])
    return matsub
def invSubBytes(mat,):
    matsub = [[],[],[],[]]
    for i in range(len(mat)):
        line = mat[i]
        for j in line:
            x = j[:1]
            y = j[1:]
            matsub[i].append(INV_SBOX[int(x,16)][int(y,16)])
    return matsub
```

```python
def shiftRows(mat):
    shift = [1,2,3]
    for i in shift:
        for j in range(i):
            elt = mat[i].pop(0)
            mat[i].append(elt)
    return mat
def invShiftRows(mat):
    shift = [1,2,3]
    for i in shift:
        for j in range(i):
            elt = mat[i].pop(3)
            mat[i].insert(0, elt)
    return mat
```

```python
def mixColumn(mat):
    mixed = [[],[],[],[]]
    mix = [
        ["02", "03", "01", "01"],
        ["01", "02", "03", "01"],
        ["01", "01", "02", "03"],
        ["03", "01", "01", "02"]
    ]
    for i in range(4):
        xora = []
        for j in range(4):
            m1 = galois_multiplication(mat[0][i], mix[j][0])
            m2 = galois_multiplication(mat[1][i], mix[j][1])
            m3 = galois_multiplication(mat[2][i], mix[j][2])
            m4 = galois_multiplication(mat[3][i], mix[j][3])
            a  = xor(m1,m2,m3,m4)
            xora.append(a)
        mixed = appendColumn(xora, mixed)
    return mixed
def invMixColumn(mat):
    mix = [
        ["0e", "0b", "0d", "09"],
        ["09", "0e", "0b", "0d"],
        ["0d", "09", "0e", "0b"],
        ["0b", "0d", "09", "0e"]
    ]
    mixed = [[],[],[],[]]
    for i in range(4):
        xora = []
        for j in range(4):
            m1 = galois_multiplication(mat[0][i], mix[j][0])
            m2 = galois_multiplication(mat[1][i], mix[j][1])
            m3 = galois_multiplication(mat[2][i], mix[j][2])
            m4 = galois_multiplication(mat[3][i], mix[j][3])
            a  = xor(m1,m2,m3,m4)
            xora.append(a)
        mixed = appendColumn(xora, mixed)
    return mixed
```

```python
def addRoundKey(mat, key):
    roundmat = [[],[],[],[]]
    for i in range(4):
        m1 = xor(mat[0][i], key[0][i],"","")
        m2 = xor(mat[1][i], key[1][i],"","")
        m3 = xor(mat[2][i], key[2][i],"","")
        m4 = xor(mat[3][i], key[3][i],"","")
        roundmat = appendColumn([m1,m2,m3,m4], roundmat)
    return roundmat
```

```python
def keyExpansion(key):
    if len(key[0]) == 8:
        okay = [[],[],[],[]]
        okay = appendMat(key, okay)
        for i in range(7):
            n_key = [[],[],[],[]]
            l1_line = [key[1][7], key[2][7], key[3][7], key[0][7]]
            l1_prime = keySubBytes(l1_line)
            rcon = RCON2[i]
            first_line = [key[0][0], key[1][0], key[2][0], key[3][0]]
            first_next_line = multi_xor(first_line,rcon, l1_prime)
            n_key = appendColumn(first_next_line, n_key)
            for j in range(1,8):
                prev_k = [n_key[0][j-1], n_key[1][j-1], n_key[2][j-1], n_key[3][j-1]]
                key_i = [key[0][j], key[1][j], key[2][j], key[3][j]]
                if j == 4:
                    prev_k = keySubBytes(prev_k)
                    n_key = appendColumn(multi_xor(key_i,prev_k,""), n_key)
                else:
                    n_key = appendColumn(multi_xor(key_i,prev_k,""), n_key)
            key = n_key
            okay = appendMat(n_key, okay)
        return subdivide(okay)
    elif len(key[0]) == 6:
        okay = [[],[],[],[]]
        okay = appendMat(key, okay)
        for i in range(8):
            n_key = [[],[],[],[]]
            l1_line = [key[1][5], key[2][5], key[3][5], key[0][5]]
            l1_prime = keySubBytes(l1_line)
            rcon = RCON2[i]
            first_line = [key[0][0], key[1][0], key[2][0], key[3][0]]
            first_next_line = multi_xor(first_line,rcon, l1_prime)
            n_key = appendColumn(first_next_line, n_key)
            for j in range(1,6):
                prev_k = [n_key[0][j-1], n_key[1][j-1], n_key[2][j-1], n_key[3][j-1]]
                key_i = [key[0][j], key[1][j], key[2][j], key[3][j]]
                n_key = appendColumn(multi_xor(key_i,prev_k,""), n_key)
            key = n_key
            okay = appendMat(n_key, okay)
        return subdivide(okay)
    elif len(key[0]) == 4:
        dicokey = {}
        dicokey["k0"] = key
        for i in range(10):
            n_key = [[],[],[],[]]
            l1_line = [key[1][3], key[2][3], key[3][3], key[0][3]]
            l1_prime = keySubBytes(l1_line)
            rcon = RCON2[i]
            first_line = [key[0][0], key[1][0], key[2][0], key[3][0]]
            first_next_line = multi_xor(first_line,rcon, l1_prime)
            n_key = appendColumn(first_next_line, n_key)
            for j in range(1,4):
                prev_k = [n_key[0][j-1], n_key[1][j-1], n_key[2][j-1], n_key[3][j-1]]
                key_i = [key[0][j], key[1][j], key[2][j], key[3][j]]
                n_key = appendColumn(multi_xor(key_i,prev_k,""), n_key)
            key = n_key
            dicokey["k"+str(i+1)] = n_key
        return dicokey
    else:
        return "Clé de chiffrement/déchiffrement non reconnue!!!"
```

```python
def genKey(size):
    if size == 256:
        ckey = [[hex(randint(0,15))[2:] + hex(randint(0,15))[2:] for i in range(8)] for j in range(4)]
        return ckey
    elif size == 192:
        ckey = [[hex(randint(0,15))[2:] + hex(randint(0,15))[2:] for i in range(6)] for j in range(4)]
        return ckey
    elif size == 128:
        ckey = [[hex(randint(0,15))[2:] + hex(randint(0,15))[2:] for i in range(4)] for j in range(4)]
        return ckey
```

```python
def preCipher(message):
    toHex = lambda x:"".join([hex(ord(c))[2:].zfill(2) for c in x])
    new_m = toHex(message)
    new_m = new_m if (len(new_m) % 32) == 0 else new_m + toHex(int((32 - (len(new_m) % 32))/2) * "{")
    new_mm = split_in(new_m, 2)
    new_mat = []
    matrix = [[],[],[],[]]
    for j in range(0, len(new_mm),4):
        if (j % 16) == 0 and j != 0:
            new_mat.append(matrix)
            matrix = [[],[],[],[]]
        matrix[0].append(new_mm[j])
        matrix[1].append(new_mm[j+1])
        matrix[2].append(new_mm[j+2])
        matrix[3].append(new_mm[j+3])
        if (len(new_mm) - 4) == j:
            new_mat.append(matrix)
    return new_mat
```

```python
def afterCipher(messblock):
    message = ""
    for mess in messblock:
        for i in range(4):
            message += mess[0][i]
            message += mess[1][i]
            message += mess[2][i]
            message += mess[3][i]
    message = toStr(message)
    return message
```

```python
def aes_encrypt(message , enc_size, key):
    mes_block = preCipher(message)
    subkeys = keyExpansion(key)
    if enc_size == 128:
        enc_block = []
        for mes in mes_block:
            state = addRoundKey(mes, subkeys["k0"])
            for i in range(1,10):
                state = addRoundKey(mixColumn(shiftRows(subBytes(state))), subkeys["k"+str(i)])
            state = addRoundKey(shiftRows(subBytes(state)), subkeys["k10"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 192:
        enc_block = []
        for mes in mes_block:
            state = addRoundKey(mes, subkeys["k0"])
            for i in range(1,12):
                state = addRoundKey(mixColumn(shiftRows(subBytes(state))), subkeys["k"+str(i)])
            state = addRoundKey(shiftRows(subBytes(state)), subkeys["k12"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 256:
        enc_block = []
        for mes in mes_block:
            state = addRoundKey(mes, subkeys["k0"])
            for i in range(1,14):
                state = addRoundKey(mixColumn(shiftRows(subBytes(state))), subkeys["k"+str(i)])
            state = addRoundKey(shiftRows(subBytes(state)), subkeys["k14"])
            enc_block.append(state)
        return afterCipher(enc_block)
    else:
        return "unrecognized encryption size"
```

```python
def aes_decrypt(message , enc_size, key):
    mes_block = preCipher(message)
    subkeys = keyExpansion(key)
    if enc_size == 128:
        enc_block = []
        for mes in mes_block:
            state = invSubBytes(invShiftRows(addRoundKey(mes, subkeys["k10"])))
            for i in range(1,10):
                state = invSubBytes(invShiftRows(invMixColumn(addRoundKey(state,subkeys["k"+str(10-i)]))))
            state = addRoundKey(state, subkeys["k0"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 192:
        enc_block = []
        for mes in mes_block:
            state = invSubBytes(invShiftRows(addRoundKey(mes, subkeys["k12"])))
            for i in range(1,12):
                state = invSubBytes(invShiftRows(invMixColumn(addRoundKey(state,subkeys["k"+str(12-i)]))))
            state = addRoundKey(state, subkeys["k0"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 256:
        enc_block = []
        for mes in mes_block:
            state = invSubBytes(invShiftRows(addRoundKey(mes, subkeys["k14"])))
            for i in range(1,14):
                state = invSubBytes(invShiftRows(invMixColumn(addRoundKey(state,subkeys["k"+str(14-i)]))))
            state = addRoundKey(state, subkeys["k0"])
            enc_block.append(state)
        return afterCipher(enc_block)
    else:
        return "unrecognized encryption size"
```
