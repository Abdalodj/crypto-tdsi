```python
def rotWord(word):
    return [word[1], word[2], word[3], word[0]]


def keySubBytes(mat):
    matsub = []
    for j in mat:
        x = j[:1]
        y = j[1:]
        matsub.append(SBOX[int(x,16)][int(y,16)])
    return matsub

def appendColumn(mat,col_mat):
    col_mat[0].append(mat[0])
    col_mat[1].append(mat[1])
    col_mat[2].append(mat[2])
    col_mat[3].append(mat[3])
    return col_mat


def appendMat(mat,col_mat):
    for i in range(len(mat[0])):
        col_mat[0].append(mat[0][i])
        col_mat[1].append(mat[1][i])
        col_mat[2].append(mat[2][i])
        col_mat[3].append(mat[3][i])
    return col_mat


def hex_bin(string):
    return "".join(
        reversed([i + j for i, j in zip(*[["{0:04b}".format(int(c, 16)) for c in reversed("0" + string)][n::2] for n
                                          in [1, 0]])]))

def toStr(s):
    return s and chr(int(s[:2], base=16)) + toStr(s[2:]) or ''


def bin_hex(string):
    return hex(int(string, 2))


def xor(a, b, c, d):
    e = hex(int(a,16) ^ int(b,16) ^ (int(c,16) if c != "" else 0) ^ (int(d,16) if d != "" else 0))
    return "0"+e[2:] if len(e[2:])==1 else e[2:]


def multi_xor(col1, col2, col3):
    new_col = []
    if isinstance(col3, list):
        for x,y,z in zip(col1, col2, col3):
            new_col.append(xor(x,y,z,""))
        return new_col
    else:
        for x,y in zip(col1, col2):
            new_col.append(xor(x,y,"",""))
        return new_col


def subdivide(matrix):
    dico = {}
    id_k = 0
    n_key = [[],[],[],[]]
    counter = 0
    for i in range(len(matrix[0])):
        counter += 1
        n_key[0].append(matrix[0][i])
        n_key[1].append(matrix[1][i])
        n_key[2].append(matrix[2][i])
        n_key[3].append(matrix[3][i])
        if counter == 4 and id_k < 15:
            dico["k"+str(id_k)] = n_key
            id_k += 1
            n_key = [[],[],[],[]]
            counter = 0
    return dico


def int_bin(integer):
    return '{0:04b}'.format(integer)


def str_bin(string):
    return "".join(f"{ord(i):08b}" for i in string)


def bin_str(string):
    return ''.join(chr(int(string[i:i+8], 2)) for i in range(0, len(string), 8))


def split_in(block, char=8):
    return [block[i:i + char] for i in range(0, len(block), char)]


def get_first_and_last_bit(bits6):
    two_bits = bits6[0] + bits6[-1]
    return int(two_bits, 2)


def get_middle_four_bit(bits6):
    four_bits = bits6[1:5]
    return int(four_bits, 2)


def split_block(block):
    k1 = ""
    k2 = ""
    for i in range(len(block)):
        if i < (len(block) / 2):
            k1 += block[i]
        else:
            k2 += block[i]
    return k1, k2
```

```python
def genKey_system(cipher_type, size):
    if cipher_type == "AES":
        if size == 256:
            ckey = [[hex(randint(0,15))[2:] + hex(randint(0,15))[2:] for i in range(8)] for j in range(4)]
            return ckey
        elif size == 192:
            ckey = [[hex(randint(0,15))[2:] + hex(randint(0,15))[2:] for i in range(6)] for j in range(4)]
            return ckey
        elif size == 128:
            ckey = [[hex(randint(0,15))[2:] + hex(randint(0,15))[2:] for i in range(4)] for j in range(4)]
            return ckey
    elif cipher_type == "DES":
        ckey = [hex(randint(0,15))[2:] for i in range(16)]
        return "".join(ckey)
    else:
        print("Type de chiffrement non reconnu")
        return 0
```

```python
def shiftRows(mat):
    shift = [1,2,3]
    for i in shift:
        for j in range(i):
            elt = mat[i].pop(0)
            mat[i].append(elt)
    return mat
def invShiftRows(mat):
    shift = [1,2,3]
    for i in shift:
        for j in range(i):
            elt = mat[i].pop(3)
            mat[i].insert(0, elt)
    return mat
```

```python
def galois_multiplication(a, b):
    """Galois multiplication of 8 bit characters a and b."""
    a = int(a,16)
    b = int(b,16)
    p = 0
    for counter in range(8):
        if b & 1: p ^= a
        hi_bit_set = a & 0x80
        a <<= 1
        # keep a 8 bit
        a &= 0xFF
        if hi_bit_set:
            a ^= 0x1b
        b >>= 1
    return str(hex(p))[2:]


def mixColumn(mat):
    mixed = [[],[],[],[]]
    for i in range(4):
        xora = []
        for j in range(4):
            m1 = galois_multiplication(mat[0][i], mix[j][0])
            m2 = galois_multiplication(mat[1][i], mix[j][1])
            m3 = galois_multiplication(mat[2][i], mix[j][2])
            m4 = galois_multiplication(mat[3][i], mix[j][3])
            a  = xor(m1,m2,m3,m4)
            xora.append(a)
        mixed = appendColumn(xora, mixed)
    return mixed
def invMixColumn(mat): 
    mixed = [[],[],[],[]]
    for i in range(4):
        xora = []
        for j in range(4):
            m1 = galois_multiplication(mat[0][i], mixInv[j][0])
            m2 = galois_multiplication(mat[1][i], mixInv[j][1])
            m3 = galois_multiplication(mat[2][i], mixInv[j][2])
            m4 = galois_multiplication(mat[3][i], mixInv[j][3])
            a  = xor(m1,m2,m3,m4)
            xora.append(a)
        mixed = appendColumn(xora, mixed)
    return mixed
```

```python
def preCipher(message):
    toHex = lambda x:"".join([hex(ord(c))[2:].zfill(2) for c in x])
    new_m = toHex(message)
    new_m = new_m if (len(new_m) % 32) == 0 else new_m + toHex(int((32 - (len(new_m) % 32))/2) * "{")
    new_mm = split_in(new_m, 2)
    new_mat = []
    matrix = [[],[],[],[]]
    for j in range(0, len(new_mm),4):
        if (j % 16) == 0 and j != 0:
            new_mat.append(matrix)
            matrix = [[],[],[],[]]
        matrix[0].append(new_mm[j])
        matrix[1].append(new_mm[j+1])
        matrix[2].append(new_mm[j+2])
        matrix[3].append(new_mm[j+3])
        if (len(new_mm) - 4) == j:
            new_mat.append(matrix)
    return new_mat
```

```python
def afterCipher(messblock):
    message = ""
    for mess in messblock:
        for i in range(4):
            message += mess[0][i]
            message += mess[1][i]
            message += mess[2][i]
            message += mess[3][i]
    message = toStr(message)
    return message
```

```python
def aes_encrypt(message , enc_size, key):
    mes_block = preCipher(message)
    subkeys = keyExpansion(key)
    if enc_size == 128:
        enc_block = []
        for mes in mes_block:
            state = addRoundKey(mes, subkeys["k0"])
            for i in range(1,10):
                state = addRoundKey(mixColumn(shiftRows(subBytes(state))), subkeys["k"+str(i)])
            state = addRoundKey(shiftRows(subBytes(state)), subkeys["k10"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 192:
        enc_block = []
        for mes in mes_block:
            state = addRoundKey(mes, subkeys["k0"])
            for i in range(1,12):
                state = addRoundKey(mixColumn(shiftRows(subBytes(state))), subkeys["k"+str(i)])
            state = addRoundKey(shiftRows(subBytes(state)), subkeys["k12"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 256:
        enc_block = []
        for mes in mes_block:
            state = addRoundKey(mes, subkeys["k0"])
            for i in range(1,14):
                state = addRoundKey(mixColumn(shiftRows(subBytes(state))), subkeys["k"+str(i)])
            state = addRoundKey(shiftRows(subBytes(state)), subkeys["k14"])
            enc_block.append(state)
        return afterCipher(enc_block)
    else:
        return "unrecognized encryption size"
```

> <h3 style="color : #B284BE; font-size: 30px; text-align: center">FONCTION DE DECHIFFREMENT AES<h3>

```python
def aes_decrypt(message , enc_size, key):
    mes_block = preCipher(message)
    subkeys = keyExpansion(key)
    if enc_size == 128:
        enc_block = []
        for mes in mes_block:
            state = invSubBytes(invShiftRows(addRoundKey(mes, subkeys["k10"])))
            for i in range(1,10):
                state = invSubBytes(invShiftRows(invMixColumn(addRoundKey(state,subkeys["k"+str(10-i)]))))
            state = addRoundKey(state, subkeys["k0"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 192:
        enc_block = []
        for mes in mes_block:
            state = invSubBytes(invShiftRows(addRoundKey(mes, subkeys["k12"])))
            for i in range(1,12):
                state = invSubBytes(invShiftRows(invMixColumn(addRoundKey(state,subkeys["k"+str(12-i)]))))
            state = addRoundKey(state, subkeys["k0"])
            enc_block.append(state)
        return afterCipher(enc_block)
    elif enc_size == 256:
        enc_block = []
        for mes in mes_block:
            state = invSubBytes(invShiftRows(addRoundKey(mes, subkeys["k14"])))
            for i in range(1,14):
                state = invSubBytes(invShiftRows(invMixColumn(addRoundKey(state,subkeys["k"+str(14-i)]))))
            state = addRoundKey(state, subkeys["k0"])
            enc_block.append(state)
        return afterCipher(enc_block)
    else:
        return "unrecognized encryption size"
```

> <h3 style="color : #00008b; font-size: 30px; text-align: center">1. l’agorithme d’Euclide pour le calcul du PGCD<h3>

```python
def euclidePGCD(a,b):
    while b != 0:
        r = a % b
        a = b
        b = r
    return a
```

> <h3 style="color : #00008b; font-size: 30px; text-align: center">2. l’agorithme d’Euclide étendu<h3>

```python
def euclideEtendu(a, b):
    r1 = a 
    r2 = b
    u1 = 1
    v1 = 0
    u2 = 0
    v2 = 1
    while r > 0:
        q = r1 // r2
        r3 = r1 
        u3 = u1 
        v3 = v1
        r1 = r2 
        u1 = u2 
        v1 = v2
        r2 = r3 - q*r2 
        u2 = u3 - q*u2 
        v2 = v3 - q*v2
    return u1, v1, r1
```

> <h3 style="color : #00008b; font-size: 30px; text-align: center">4. l’algorithme de puissance (modulaire) rapide<h3>

```python
def puissanceRapide(a, m, n):
    b = 1
    if m == 0:
        return b
    else:
        ga = a
        bin_m = bin(m)[2:]
        for k in bin_m:
            ga = (ga**2) % n
            if k == 1:
                b = ga * b
        return b
```
